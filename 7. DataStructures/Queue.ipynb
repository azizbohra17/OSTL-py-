```python
class q:
 def __init__(self):
  self.queue = []
 def push(self,item):
  self.queue.append(item)

 def pop(self):
  print(self.queue.pop(0))

 def isEmpty(self):
  if(self.queue == []):
   print("queue is empty")
  else:
   print("queue is not empty")
 def size(self):
  print(len(self.queue))
 def display(self):
  for a in range(len(self.queue)):
   print(self.queue[a])

ob = q()
while 1:
 print("1 to Push")
 print("2 to Pop")
 print("3 to find Length")
 print("4 to find if its empty or not")
 print("5 to display")
 print("6 to exit")
 x = input()
 if(x=="1"):
  t = input("Enter item")
  ob.push(t) 
 elif(x=="2"):
  ob.pop() 
 elif(x=="3"):
  ob.size() 
 elif(x=="4"):
  ob.isEmpty() 
 elif(x=="5"):
  ob.display() 
 elif(x=="6"):
  break
```
